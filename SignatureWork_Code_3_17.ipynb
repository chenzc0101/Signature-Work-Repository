```python
look_back = 1
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))


# Create function for creating dataset with look back
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back):
        a = dataset[i, :]
        dataX.append(a)
        dataY.append(dataset[(i+1):(i+1+look_back), 0])
    print(len(dataY))
    return dataX, np.array(scaler.fit_transform(dataX)), dataY, np.array(scaler.fit_transform(dataY))

# # Generate dataset for trainX, trainY, testX, testY
# #trainX_untr, trainX, trainY_untr, trainY = create_dataset(train_deno.values, look_back)
# trainX_untr, trainX, trainY_untr, trainY = create_dataset(train.values, look_back)
# devX_untr, devX, devY_untr, devY = create_dataset(dev.values, look_back)
# testX_untr, testX, testY_untr, testY = create_dataset(test.values, look_back)
```

```python
from sklearn.metrics import mean_squared_error
def acc_metric(true_value, predicted_value):
    acc_met = 0.0
    m = len(true_value)
    for i in range(m):
        acc_met += mean_squared_error(true_value[i], predicted_value[i])
    acc_met /= m
    return np.sqrt(acc_met)

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

## Part4. LSTM Model

### 4.1 Build LSTM Function

```python
#

def LSTM_Train(asset):

  train_size = int(len(lstm_pair_data_[asset]) * 0.8)
  dev_size = int((len(lstm_pair_data_[asset]) - train_size) * 0.5) - 30
  test_size = len(lstm_pair_data_[asset]) - train_size - dev_size
  train, dev, test = lstm_pair_data_[asset][0:train_size], lstm_pair_data_[asset][train_size:train_size + dev_size], lstm_pair_data_[asset][train_size + dev_size:len(lstm_pair_data_[asset])]

  trainX_untr, trainX, trainY_untr, trainY = create_dataset(train.values, look_back)
  devX_untr, devX, devY_untr, devY = create_dataset(dev.values, look_back)
  testX_untr, testX, testY_untr, testY = create_dataset(test.values, look_back)

  # Reshape X for model training
  trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
  devX = np.reshape(devX, (devX.shape[0], 1, devX.shape[1]))
  testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
  # Running the LSTM model
  model = Sequential()
  model.add(LSTM(256, input_shape=(trainX.shape[1], trainX.shape[2]), return_sequences = True))
  model.add(LSTM(256))
  model.add(Dropout(0.2))
  model.add(Dense(look_back))
  model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
  history = model.fit(trainX, trainY, epochs=200, batch_size=20, validation_data=(devX, devY), verbose=0, shuffle=True)

  # Result Accuracy Assesment
  
  yhat = model.predict(testX)
  #mse
  mse_score = acc_metric(testY, yhat)

  from sklearn.metrics import r2_score
  #r2
  r2_score = r2_score(testY,yhat)

  # print(asset+ 'score :' + mse_score + '+' + r2_score)

  return yhat, mse_score, r2_score


```

## Part5. ARIMA Model


### 5.1 Preparation for ARIMA

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.vector_ar.var_model import VAR

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error

from sklearn.metrics import r2_score
```

### 5.2 ARIME model function

```python
def ARIMA_Train(asset):
  yhat_ARIMA = []
  yhat_ARIMA_mse = []
  data = lstm_pair_data_[asset][asset+'_close'].values

  train_size = int(len(lstm_pair_data_[asset]) * 0.8)
  dev_size = int((len(lstm_pair_data_[asset]) - train_size) * 0.5) - 30
  test_size = len(lstm_pair_data_[asset]) - train_size - dev_size
  train, dev, test = lstm_pair_data_[asset][0:train_size], lstm_pair_data_[asset][train_size:train_size + dev_size], lstm_pair_data_[asset][train_size + dev_size:len(lstm_pair_data_[asset])]

  trainX_untr, trainX, trainY_untr, trainY = create_dataset(train.values, look_back)
  devX_untr, devX, devY_untr, devY = create_dataset(dev.values, look_back)
  testX_untr, testX, testY_untr, testY = create_dataset(test.values, look_back)

  # Reshape X for model training
  trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
  devX = np.reshape(devX, (devX.shape[0], 1, devX.shape[1]))
  testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

  for i in range(train_size+dev_size, len(lstm_pair_data_[asset])-1):
      model = ARIMA(data[:i], order=(1,0,0))
      model_fit = model.fit(disp=0)
      temp = []
      forecast = (model_fit.forecast(steps=look_back)[0])
      yhat_ARIMA.append(forecast[0])
      for j in range(len(forecast)):
          temp.append(forecast[j])
      yhat_ARIMA_mse.append(np.array(temp))
  
  mse = acc_metric(testY_untr, yhat_ARIMA_mse)

  from sklearn.metrics import r2_score
  r2 = r2_score(testY_untr,yhat_ARIMA_mse)

  return yhat_ARIMA_mse, mse, r2

```

## Part6. VAR (Vector Auto-Regression Model)

### 6.1 VAR preparation and function


```python
def VAR_Train(asset):

  yhat_VAR_mse = []
  yhat_VAR = []
  data = lstm_pair_data_[asset].values

  train_size = int(len(lstm_pair_data_[asset]) * 0.8)
  dev_size = int((len(lstm_pair_data_[asset]) - train_size) * 0.5) - 30
  test_size = len(lstm_pair_data_[asset]) - train_size - dev_size
  train, dev, test = lstm_pair_data_[asset][0:train_size], lstm_pair_data_[asset][train_size:train_size + dev_size], lstm_pair_data_[asset][train_size + dev_size:len(lstm_pair_data_[asset])]

  trainX_untr, trainX, trainY_untr, trainY = create_dataset(train.values, look_back)
  devX_untr, devX, devY_untr, devY = create_dataset(dev.values, look_back)
  testX_untr, testX, testY_untr, testY = create_dataset(test.values, look_back)

  # Reshape X for model training
  trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
  devX = np.reshape(devX, (devX.shape[0], 1, devX.shape[1]))
  testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

  for i in range(train_size+dev_size, len(lstm_pair_data_[asset])-1):
      model = VAR(data[:i,:])
      model_fit = model.fit()
      temp = []
      forecast = model_fit.forecast(model_fit.y, steps=look_back)[:,0]
      yhat_VAR.append(forecast[0])
      for j in range(len(forecast)):
          temp.append(forecast[j])
      yhat_VAR_mse.append(np.array(temp))
  

  mse = acc_metric(testY_untr, yhat_VAR_mse)

  from sklearn.metrics import r2_score
  r2 = r2_score(testY_untr,yhat_VAR_mse)

  return yhat_VAR_mse, mse, r2

```
